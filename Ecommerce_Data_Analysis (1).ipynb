```python
df_clean= df.copy()
```

```python
df_clean.isnull().sum()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.dtypes
```

```python
df_clean['Transaction_Date'] = pd.to_datetime(
    df_clean['Transaction_Date']
)
```

```python
df_clean.dtypes
```

```python
df_clean['Year'] = df_clean['Transaction_Date'].dt.year
```

```python
df_clean[['Transaction_Date', 'Year']].head()  
```

```python
df_clean['Month'] = df_clean['Transaction_Date'].dt.month_name()
```

```python
df_clean[['Transaction_Date', 'Month']].head()
```

```python
df_clean['Day'] = df_clean['Transaction_Date'].dt.day_name()
```

```python
df_clean[['Transaction_Date', 'Day']].head()
```

```python
df_clean['Spending_Category'] = pd.cut(
    df_clean['Purchase_Amount'],
    bins=[0, 200, 500, 1000],
    labels=['Low', 'Medium', 'High']
)
```

```python
df_clean[
    ['Purchase_Amount',
     'Spending_Category']
].head()
```

```python
df_clean['Spending_Category'].value_counts()
```

```python
sports = df_clean[
    df_clean['Product_Category'] == 'Sports'
]

sports.head()
```

```python
df_clean.to_csv(
    'ecommerce_cleaned.csv',
    index=False
)
```

```python
df_clean.head()
```

```python
sns.set_style('whitegrid')
```

```python
category_revenue = (
    df_clean.groupby('Product_Category')['Purchase_Amount']
    .sum()
    .sort_values(ascending=False)
)
category_revenue
```

```python
plt.figure(figsize=(10,6))

category_revenue.plot(kind='bar')
plt.title('Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.show()
```

```python
country_revenue = (
    df_clean.groupby('Country')['Purchase_Amount']
    .sum()
    .sort_values(ascending=False)
)

country_revenue
```

```python
plt.figure(figsize=(12,6))

country_revenue.plot(kind='bar')

plt.title('Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Revenue')

plt.show()
```

```python
payment_counts = (
    df_clean['Payment_Method']
    .value_counts()
)

payment_counts
```

```python
plt.figure(figsize=(8,8))

payment_counts.plot(kind='pie', autopct='%1.1f%%')

plt.title('Payment Method Distribution')
plt.ylabel('')

plt.show()
```

```python
spending_counts = (
    df_clean['Spending_Category']
    .value_counts()
)

spending_counts
```

```python
plt.figure(figsize=(8,6))

spending_counts.plot(kind='bar')

plt.title('Spending Categories')
plt.xlabel('Category')
plt.ylabel('Transactions')

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(
    df_clean['Age'],
    bins=10
)

plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
numerical_columns = [
    'Age',
    'Purchase_Amount'
]

correlation = (
    df_clean[numerical_columns]
    .corr()
)

correlation
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True
)

plt.title('Correlation Matrix')

plt.show()
```

```python
df_clean['Year_Month'] = (
    df_clean['Transaction_Date']
    .dt.to_period('M')
)
```

```python
monthly_revenue = (
    df_clean
    .groupby('Year_Month')['Purchase_Amount']
    .sum()
)

monthly_revenue
```

```python
plt.figure(figsize=(12,6))

monthly_revenue.plot(
    kind='line',
    marker='o'
)

plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.show()
```

```python
top_customers = (
    df_clean
    .groupby('User_Name')['Purchase_Amount']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_customers
```

```python
plt.figure(figsize=(12,6))

top_customers.plot(kind='bar')

plt.title('Top 10 Customers by Revenue')
plt.xlabel('Customer')
plt.ylabel('Revenue')

plt.xticks(rotation=45)

plt.show()
```

```python
spending_revenue = (
    df_clean
    .groupby('Spending_Category')['Purchase_Amount']
    .sum()
)
spending_revenue
```

```python
plt.figure(figsize=(8,6))

spending_revenue.plot(
    kind='bar'
)

plt.title('Revenue by Spending Category')
plt.xlabel('Category')
plt.ylabel('Revenue')

plt.show()
```
